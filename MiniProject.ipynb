# Mini Project

## Research Question
# How has the total population of each continent changed between 1970 and 2022 and which experienced the highest growth rate?

## Deliverables
# Source- https://www.kaggle.com/datasets/iamsouravbanerjee/world-population-dataset/data
# Description- Population data by continent 1970-2022
# Time Period- 1970-2022
# Reason for Choice- Simple data set, also just interesting to see

```python
import boto3
import pandas as pd

bucket_name = "ia-340-epage"  
file_key = "world_population.csv"  

# Create S3 client
s3 = boto3.client('s3')

# Load dataset
response = s3.get_object(Bucket=bucket_name, Key=file_key)

# Read CSV into pandas
df = pd.read_csv(response['Body'])

# Show a sample
df.head()
```

```python
# Inspect structure
df.info()
df.describe(include='all')
```

```python
import pandas as pd

# Drop duplicates (none expected, but safe)
df = df.drop_duplicates()

# Keep only relevant columns
df = df[['Country/Territory', 'Continent', 
         '2022 Population', '2020 Population', '2015 Population',
         '2010 Population', '2000 Population', '1990 Population',
         '1980 Population', '1970 Population']]

# Rename for convenience
df = df.rename(columns={'Country/Territory': 'Country'})

# Melt the dataset to long format
df_long = df.melt(id_vars=['Country', 'Continent'],
                  var_name='Year',
                  value_name='Population')

# Clean the Year column (remove " Population" and convert to int)
df_long['Year'] = df_long['Year'].str.replace(' Population', '').astype(int)

# Convert population to numeric
df_long['Population'] = pd.to_numeric(df_long['Population'], errors='coerce')

# Drop missing or invalid entries
df_long = df_long.dropna(subset=['Country', 'Continent', 'Year', 'Population'])
df_long = df_long[df_long['Population'] > 0]

# Check the cleaned structure
df_long.info()
df_long.head()
```

```python
# Group by continent and year, summing population
continent_pop = (
    df_long.groupby(['Continent', 'Year'])['Population']
    .sum()
    .reset_index()
)

# Calculate percent change by continent
continent_pop['Pct_Change'] = (
    continent_pop.groupby('Continent')['Population']
    .pct_change() * 100
)

continent_pop.head()
```

```python
# Query 2020 data
pop_2020 = continent_pop[continent_pop['Year'] == 2020].sort_values(by='Population', ascending=False)
pop_2020
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(10,6))

for continent in continent_pop['Continent'].unique():
    subset = continent_pop[continent_pop['Continent'] == continent]
    plt.plot(subset['Year'], subset['Population']/1e9, marker='o', label=continent)

plt.title('Population Growth by Continent (1970–2022)')
plt.xlabel('Year')
plt.ylabel('Population (Billions)')
plt.legend()
plt.grid(True)
plt.show()
```

```python
# Calculate average percent change per continent
avg_growth = (
    continent_pop.groupby('Continent')['Pct_Change']
    .mean()
    .reset_index()
    .sort_values(by='Pct_Change', ascending=False)
)

plt.figure(figsize=(8,5))
plt.bar(avg_growth['Continent'], avg_growth['Pct_Change'], color='skyblue', edgecolor='black')
plt.title('Average Annual Population Growth Rate by Continent (1970–2022)')
plt.xlabel('Continent')
plt.ylabel('Average % Growth Rate')
plt.show()

avg_growth
```
